### Let's Change Object to Numeric data

```python
from sklearn.preprocessing import LabelEncoder
for colums in df.select_dtypes(include= ['object']).columns:
    le = LabelEncoder()
    df[colums]=le.fit_transform(df[colums])
```

```python
df.head(10)
```

### I've seen a lot of zeros, let's find out what percentage of zeros in df.

```python
# Function to calculate percentage of 0s in each column
def calculate_zero_percentage(df):
    percentages = {}
    for col in df.columns:
        # Count the number of 0s in the column
        zero_count = (df[col] == 0).sum()
        # Calculate percentage
        percentage = (zero_count / len(df)) * 100
        percentages[col] = percentage
    return percentages

# Get the percentages for all columns
zero_percentages = calculate_zero_percentage(df)

# Display the result
for col, percentage in zero_percentages.items():
    print(f"Percentage of 0s in column '{col}': {percentage:.2f}%")
```

# Modelling

```python
from sklearn import metrics, linear_model, neighbors, model_selection
```

```python
x = df.drop(columns = ['income'])
y = df.income

x.shape, y.shape
```

### Scalling Features

```python
from sklearn import preprocessing

x_normalize = preprocessing.StandardScaler()
x_norm = x_normalize.fit_transform(x)

x_norm.shape, x.shape, y.shape
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(
    x_norm, y, test_size = 0.2, random_state = 42, stratify = y
)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

### 1- Logistic Regression (Classification)

```python
log_model = linear_model.LogisticRegression()
log_model.fit(x_train, y_train)
```

```python
y1_pred = log_model.predict(x_test)
```

```python
acc = metrics.accuracy_score(y_test, y1_pred)
print(f'Accuracy = {int(acc*100)}%')
```

### 2- KNN Algorithm (classification)

```python
knn_model = neighbors.KNeighborsClassifier(n_neighbors=15)
knn_model.fit(x_train, y_train)
```

```python
y2_pred = knn_model.predict(x_test)
```

```python
acc = metrics.accuracy_score(y_test, y2_pred)
print(f'Accuracy = {int(acc*100)}%')
```
